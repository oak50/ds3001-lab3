3. Use Numpy's .reshape() method to covert the training and testing data from a matrix into an vector of features. So, X_test[index].reshape((1,784)) will convert the index-th element of X_test into a 28 x 28 = 784-length row vector of values, rather than a matrix. Turn X_train into an N x 784 matrix X that is suitable for scikit-learn's kNN classifier where N is the number of observations and 784 = 28 * 28 (you could use, for example, a for loop).

```python
# code generatied with help from chat gpt:


# Load MNIST dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Control the way data loads
reload = 1  # Set to 1 to reshape and save the data

# Define the directory and file paths
data_dir = os.path.abspath('./data')
train_file = os.path.join(data_dir, 'Z_train.csv')
test_file = os.path.join(data_dir, 'Z_test.csv')

# Create the directory if it doesn't exist
os.makedirs(data_dir, exist_ok=True)
print(f"Data directory path: {data_dir}")

if reload == 1:  # If reload is 1, do the reshaping process
    print("Reloading and reshaping data...")
    # Check shape of data before reshaping
    print(f"Original X_train shape: {X_train.shape}")
    print(f"Original X_test shape: {X_test.shape}")
    
    try:
        # Reshape the training and test data into a vectorized form
        Z_train = pd.DataFrame(X_train.reshape(-1, 784))
        Z_test = pd.DataFrame(X_test.reshape(-1, 784))
        print("Data reshaped successfully.")
    except Exception as e:
        print(f"Reshaping failed: {e}")
        Z_train, Z_test = None, None

    # Attempt to save the reshaped data
    if Z_train is not None and Z_test is not None:
        try:
            Z_train.to_csv(train_file, index=False)
            print(f"Successfully saved training data to {train_file}")
        except Exception as e:
            print(f"Failed to save training data: {e}")

        try:
            Z_test.to_csv(test_file, index=False)
            print(f"Successfully saved test data to {test_file}")
        except Exception as e:
            print(f"Failed to save test data: {e}")
else:  # If reload is not 1, just load the reshaped data
    print("Loading data from saved files...")
    if os.path.exists(train_file) and os.path.exists(test_file):
        try:
            Z_train = pd.read_csv(train_file)
            Z_test = pd.read_csv(test_file)
            print(f"Loaded training data from {train_file}")
            print(f"Loaded test data from {test_file}")
        except Exception as e:
            print(f"Failed to load saved data: {e}")
    else:
        print("Saved files not found. Please set `reload = 1` to reshape and save the data.")
```

4. Use the reshaped X_test and y_test data to create a k-nearest neighbor classifier of digit. What is the optimal number of neighbors 
? If you can't determine this, play around with different values of k for your classifier.

```python
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
```

```python

# code generated with help from chat gpt:

# Keep the full dataset intact
X_train_full = Z_train.values  # Use .values to convert DataFrame to numpy array for faster processing
X_test_full = Z_test.values
y_train_full = y_train
y_test_full = y_test

# Define a range of k values to test, keeping it reasonable to avoid excessive computation
k_bar = 20  # Choose up to 20 neighbors for now, can be increased later if needed
k_grid = np.arange(1, k_bar + 1)  # Consider k values from 1 to 20

# Store accuracy scores for each k
accuracy = np.zeros(k_bar) 

# Loop through the k values and evaluate the classifier
for k in k_grid:
    print(f"Evaluating k={k}...")  # Print progress for each k
    knn = KNeighborsClassifier(n_neighbors=k, n_jobs=-1)  # Utilize all cores for parallel processing
    knn.fit(X_train_full, y_train_full)  # Fit on the entire training set
    
    # Evaluate accuracy on the full test set
    accuracy[k-1] = knn.score(X_test_full, y_test_full)  # Store accuracy at index k-1

# Find the highest accuracy and the corresponding k value
accuracy_max = np.max(accuracy)
optimal_k = k_grid[np.argmax(accuracy)]  # Get the optimal k with highest accuracy
print(f"Optimal k: {optimal_k}, Highest Accuracy: {accuracy_max:.4f}")

# Plot accuracy vs. k values
plt.plot(k_grid, accuracy, marker='o')
plt.xlabel("k")
plt.ylabel("Accuracy")
plt.title(f"Optimal k: {optimal_k}, Highest Accuracy: {accuracy_max:.4f}")
plt.grid(True)
plt.show()
```

optimal k is 3.

To make sure you follow this, invent a way to represent a color photo in matrix form, and then describe how you could convert it into tabular data. (Hint: RGB color codes provide a method of encoding a numeric value that represents a color.)

In order to perform this task, you could create take a photo and look at the matrix of colored pixels, saving the numerical equivalents for pixel pigments in a matrix dataset. I tried to replicate this (on a small scale, not using a picture) in code using chat gpt:

```python
# created with help from chat gpt:

image_matrix = np.array([
    [[255, 0, 0], [0, 255, 0]],   # Row 1: Red, Green
    [[0, 0, 255], [255, 255, 0]],  # Row 2: Blue, Yellow
    [[255, 0, 0], [0, 0, 255]], # Row 3: Red, Blue
    [[255, 0, 0], [0, 255, 0]] # Row 4: Red, Green
])

# Step 2: Convert the 3D matrix into tabular data
# Get the height and width of the image
height, width, _ = image_matrix.shape

# Create lists to hold pixel data
pixel_data = []

# Loop through each pixel and extract RGB values
for row in range(height):
    for col in range(width):
        r, g, b = image_matrix[row, col]
        pixel_data.append([row, col, r, g, b])  # Append pixel position and RGB values

# Step 3: Convert the pixel data into a pandas DataFrame
df = pd.DataFrame(pixel_data, columns=['Row', 'Column', 'Red', 'Green', 'Blue'])

# Step 4: Display the resulting DataFrame
print("Tabular representation of the image:")
print(df)

X, Y = np.meshgrid(np.arange(width), np.arange(height))

# Plotting each color channel
color_channels = ['Red', 'Green', 'Blue']
for channel in color_channels:
    plt.figure(figsize=(6, 4))
    
    # Extract the channel values
    Z = image_matrix[:, :, color_channels.index(channel)]
    
    # Create the contour plot
    plt.contourf(X, Y, Z, cmap='viridis')
    plt.colorbar(label=channel + ' intensity')
    plt.title(f'Contour Plot of {channel} Channel')
    plt.xlabel('Column Index')
    plt.ylabel('Row Index')
    plt.grid(True)
    plt.show()
```
